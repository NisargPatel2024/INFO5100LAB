# Lab 8 Assignment: Data Visualizations using Python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading a dataset:
penguins_dataframe = pd.read_csv('PenguinsDataset/penguins_lter.csv')
```

## Line Plot

### Univariate
Visual representation of the relationship of a single column in a dataset.

```python
plt.plot(penguins_dataframe["Sample Number"], penguins_dataframe['Body Mass (g)'])
plt.title("Body mass of Penguins")
plt.xlabel("Id")
plt.ylabel("Body Mass (g)")
# plt.show()
```

## Scatter Plot

### Bivariate
Visual representation of the relationship between two variables in a dataset.

```python
plt.grid(True)
plt.scatter(penguins_dataframe["Culmen Length (mm)"], penguins_dataframe["Culmen Depth (mm)"])
plt.title("Penguins Culmen Length vs Depth")
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (mm)")
```

### Multivariate
Visual representation of the relationship between three or more columns/ variables in a dataset.

```python
# Species
# Adelie Penguin (Pygoscelis adeliae)          152
# Gentoo penguin (Pygoscelis papua)            124
# Chinstrap penguin (Pygoscelis antarctica)     68

penguin_species_color = {"Adelie Penguin (Pygoscelis adeliae)": "red", 
                      "Gentoo penguin (Pygoscelis papua)": "green", 
                      "Chinstrap penguin (Pygoscelis antarctica)":"blue"}

for species, species_color in penguin_species_color.items():
    # creating a data frame for a specific value for 'Species' column:
    species_dataframe = penguins_dataframe[penguins_dataframe["Species"] == species]
    # parameter c - color
    plt.scatter(species_dataframe["Culmen Length (mm)"], species_dataframe["Culmen Depth (mm)"], c = species_color, label=species)

plt.grid(True)
plt.title("Penguins Culmen Length vs Depth")
plt.xlabel("Culmen Length (mm)")
plt.ylabel("Culmen Depth (cm)")
plt.legend()
```

## Histograms

```python
sns.histplot(data=penguins_dataframe, x="Flipper Length (mm)")
plt.title("Penguins Flipper Length Distribution")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Count")
```

## Kernel Density Plot (KDE)

KDE plot helps to visualize the distribution of data in a dataset



```python
# Create a KDE plot
plt.figure(figsize=(10, 10))
sns.kdeplot(data=penguins_dataframe, x="Flipper Length (mm)", hue="Species", fill=True, common_norm=False, palette="crest", alpha=0.2)

# Add labels and title
plt.title("Kernel Density Plot of Flipper Length by Species")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Density")
plt.show()
```

## Pair Plot
Introduce pairwise relationships.

```python
sns.pairplot(penguins_dataframe, hue="Species")
plt.show()
```
